```python

data_osn = pd.read_csv('osn.csv')

kategori_inf = data_osn[(data_osn['Jenjang Lomba'] == 'SMA') & (data_osn['Bidang'] == 'Informatika')]

kategori_inf_reset = kategori_inf.reset_index(drop=True)


if 'Unnamed: 0' in kategori_inf_reset.columns:
    del kategori_inf_reset['Unnamed: 0']


urutan_medali = {
    'Emas': 1,
    'Perak': 2,
    'Perunggu': 3,
    'Partisipan': 4,
}


kategori_inf_reset.loc[:, 'Peringkat'] = kategori_inf_reset['Medali'].map(urutan_medali)


kategori_inf_sorted = kategori_inf_reset.sort_values(by='Peringkat').reset_index(drop=True)
kategori_inf_sorted
```

```python
# Membuat grafik garis
plt.figure(figsize=(10, 6))
plt.plot(jumlah_peserta_per_tahun.index, jumlah_peserta_per_tahun.values, marker='o', linestyle='-', color='r')
plt.title('CHART LINE PESERTA OSN BIDANG INFORMATIKA TAHUN 2009-2024')
plt.xlabel('Tahun')
plt.ylabel('Jumlah Peserta')
plt.xticks(jumlah_peserta_per_tahun.index)  
plt.grid()


plt.tight_layout()
plt.show()
```

```python
x_values = jumlah_peserta_per_tahun.index
y_values = jumlah_peserta_per_tahun.values.flatten() 

plt.figure(figsize=(10, 6))
plt.scatter(x=x_values, y=y_values, color='b', marker='o')
plt.title('DIAGRAM PENCAR PESERTA OSN BIDANG INFORMATIKA TAHUN 2009-2024')
plt.xlabel('Tahun')
plt.ylabel('Jumlah Peserta')
plt.xticks(x_values)
plt.grid()

slope, intercept = np.polyfit(x_values, y_values, 1) 

regression_line = slope * x_values + intercept

plt.plot(x_values, regression_line, color='r', linestyle='--', label='Garis Regresi')
plt.legend()

korelasi = np.corrcoef(x_values, y_values)[0][1]
plt.text(2010, max(y_values) * 0.8,
         f'Korelasi: {korelasi:.2f}', fontsize=12)

plt.tight_layout()
plt.show()
```
