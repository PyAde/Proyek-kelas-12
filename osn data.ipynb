```python

data_osn = pd.read_csv('osn.csv')

kategori_sma_inf = data_osn[(data_osn['Jenjang Lomba'] == 'SMA') & (data_osn['Bidang'] == 'Informatika')]

kategori_sma_inf_reset = kategori_sma_inf.reset_index(drop=True)


if 'Unnamed: 0' in kategori_sma_inf_reset.columns:
    del kategori_sma_inf_reset['Unnamed: 0']


urutan_medali = {
    'Emas': 1,
    'Perak': 2,
    'Perunggu': 3,
    'Peserta': 4,
}


kategori_sma_inf_reset.loc[:, 'Peringkat'] = kategori_sma_inf_reset['Medali'].map(urutan_medali)


kategori_sma_inf_sorted = kategori_sma_inf_reset.sort_values(by='Peringkat').reset_index(drop=True)
kategori_sma_inf_sorted
```

```python
kategori_sma_inf_sorted.info()
```

```python
# kategori_sma_inf_sorted

prize_tambahan_sorted = kategori_sma_inf_sorted.sort_values(by=['Peringkat', 'Prize Tambahan'], na_position='last')

prize_tambahan_sorted.drop(columns=['Peringkat'], inplace=True)
prize_tambahan_sorted.reset_index(drop=True, inplace=True)
prize_tambahan_sorted
```
